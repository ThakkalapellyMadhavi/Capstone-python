### Separating Categorical and Numerical Variables

```python
#seperating the categorical and Numerical data

def seperate_data_types(df):
    categorical=[]
    continuous=[]
    for columns in df.columns:
        if df[columns].nunique()<100:             #return the nunique values in the column 
            categorical.append(columns)         # 100 is hyperparameter
        else:
            continuous.append(columns)
    return categorical,continuous

categorical, continuous = seperate_data_types(df)

from tabulate import tabulate 
table=[categorical,continuous]
print(tabulate({"Categorical":categorical,"Continous":continuous},headers=["categorical","continuous"]))
```
